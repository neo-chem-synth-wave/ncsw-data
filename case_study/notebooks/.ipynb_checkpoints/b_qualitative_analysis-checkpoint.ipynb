```python
from matplotlib import pyplot
from matplotlib.ticker import LogLocator, NullFormatter

from numpy import arange, linspace

from pandas import DataFrame

from seaborn import heatmap, lineplot

from ncsw_data.storage.cacs.sqlite_ import CaCSSQLiteDatabase
```

```python
case_study_cacs_db_file_path = (
    "sqlite:////path/to/the/case_study_cacs_db.sqlite"
)

case_study_cacs_db = CaCSSQLiteDatabase(
    db_url=case_study_cacs_db_file_path
)

case_study_cacs_db.create_tables()
```

# Chemical Reactions

```python
select_statement_results = case_study_cacs_db.execute_select_statement(
    select_statement_text="""
        SELECT
            number_of_reactant_compounds,
            COUNT(*) AS number_of_workbench_reactions
        FROM (
            SELECT
                COUNT(workbench_compound_id) AS number_of_reactant_compounds
            FROM workbench_reaction_reactant_compound
            GROUP BY workbench_reaction_id
        )
        GROUP BY number_of_reactant_compounds
        ORDER BY number_of_reactant_compounds;
    """
).fetchall()

for select_statement_result in select_statement_results:
    print(
        "Number of workbench chemical reactions with {number_of_reactant_compounds:d} reactant compounds: {number_of_workbench_reactions:,d}".format(
            number_of_reactant_compounds=select_statement_result[0],
            number_of_workbench_reactions=select_statement_result[1]
        )
    )
```

```python
select_statement_results = case_study_cacs_db.execute_select_statement(
    select_statement_text="""
        SELECT
            number_of_reactant_compounds,
            number_of_building_block_compounds,
            COUNT(*) AS number_of_workbench_reactions
        FROM (
            SELECT
                wrrc.workbench_reaction_id,
                COUNT(*) AS number_of_reactant_compounds,
                SUM(CAST(wc.is_building_block AS INT)) AS number_of_building_block_compounds
            FROM workbench_reaction_reactant_compound AS wrrc
                JOIN workbench_compound AS wc ON wrrc.workbench_compound_id = wc.id
            GROUP BY wrrc.workbench_reaction_id
        )
        GROUP BY number_of_reactant_compounds, number_of_building_block_compounds
        ORDER BY number_of_reactant_compounds, number_of_building_block_compounds;
    """
).fetchall()

number_of_workbench_reactions_per_number_of_reactants, select_statement_data = dict(), list()

for select_statement_result in select_statement_results:
    print((
        "Number of workbench chemical reactions with {number_of_reactant_compounds:d} reactant and {number_of_building_block_compounds:d} "
        "building block compounds: {number_of_workbench_reactions:,d}"
    ).format(
        number_of_reactant_compounds=select_statement_result[0],
        number_of_building_block_compounds=select_statement_result[1],
        number_of_workbench_reactions=select_statement_result[2]
    ))

    if select_statement_result[0] not in number_of_workbench_reactions_per_number_of_reactants:
        number_of_workbench_reactions_per_number_of_reactants[select_statement_result[0]] = 0

    number_of_workbench_reactions_per_number_of_reactants[select_statement_result[0]] += select_statement_result[2]

    select_statement_data.append(
        list(select_statement_result)
    )

select_statement_dataframe = DataFrame(
    data=select_statement_data,
    columns=[
        "number_of_reactant_compounds",
        "number_of_building_block_compounds",
        "number_of_workbench_reactions",
    ]
)
```

```python
visualization_dataframe = select_statement_dataframe.groupby("number_of_reactant_compounds")["number_of_workbench_reactions"].sum().reset_index()

visualization_dataframe["cumulative_percentage_of_workbench_reactions"] = (
    100 * visualization_dataframe["number_of_workbench_reactions"].cumsum() / select_statement_dataframe["number_of_workbench_reactions"].sum()
)

pyplot.figure(
    figsize=(14, 8),
    dpi=300
)

lineplot(
    data=visualization_dataframe,
    x="number_of_reactant_compounds",
    y="cumulative_percentage_of_workbench_reactions",
    marker="o",
    color="firebrick"
)

pyplot.xlabel(
    xlabel="Number of Workbench Chemical Reaction Reactant Compounds",
    fontsize=12
)

pyplot.ylabel(
    ylabel="Cumulative Percentage of Workbench Chemical Reactions (%)",
    fontsize=12
)

pyplot.xticks(
    ticks=arange(
        visualization_dataframe["number_of_reactant_compounds"].min(),
        visualization_dataframe["number_of_reactant_compounds"].max() + 1,
        1
    )
)

pyplot.yticks(
    ticks=linspace(0, 100, 11)
)

pyplot.tick_params(
    axis="both",
    labelsize=12
)

pyplot.grid(
    visible=True,
    linestyle="--",
    alpha=0.5
)

pyplot.show()
```

```python
visualization_dataframe = DataFrame(
    data=select_statement_dataframe
)

visualization_dataframe["percentage_of_workbench_reactions"] = visualization_dataframe.apply(
    func=lambda visualization_dataframe_row: (
        100 * visualization_dataframe_row["number_of_workbench_reactions"] / number_of_workbench_reactions_per_number_of_reactants[
            visualization_dataframe_row["number_of_reactant_compounds"]
        ]
    ),
    axis=1
)

visualization_dataframe = visualization_dataframe.pivot(
    index="number_of_reactant_compounds",
    columns="number_of_building_block_compounds",
    values="percentage_of_workbench_reactions"
).fillna(0)

pyplot.figure(
    figsize=(14, 10),
    dpi=300
)

axes = heatmap(
    data=visualization_dataframe,
    cmap="Reds",
    annot=True,
    fmt=".1f",
    linewidths=0.5,
    annot_kws={
        "fontsize": 12
    } 
)

axes.collections[0].colorbar.ax.set_ylabel(
    ylabel="Percentage of Workbench Chemical Reactions (%)",
    fontsize=12
)

axes.collections[0].colorbar.ax.tick_params(
    labelsize=12
)

axes.set_aspect(
    "equal"
)

axes.set_xlabel(
    xlabel="Number of Building Block Chemical Compounds",
    fontsize=12
)

axes.set_ylabel(
    ylabel="Number of Workbench Chemical Reaction Reactant Compounds",
    fontsize=12
)

pyplot.yticks(
    rotation=0
)

pyplot.tick_params(
    axis="both",
    labelsize=12
)

pyplot.show()
```

```python
select_statement_result = case_study_cacs_db.execute_select_statement(
    select_statement_text="""
        SELECT
            COUNT(*) AS number_of_workbench_reactions
        FROM workbench_reaction_product_compound AS wrpc
            JOIN workbench_compound AS wc ON wrpc.workbench_compound_id = wc.id
        WHERE wc.is_building_block is TRUE;
    """
).fetchone()

print(
    "Number of workbench chemical reactions with a building block product compound: {number_of_workbench_reactions:,d}".format(
        number_of_workbench_reactions=select_statement_result[0]
    )
)
```

```python
select_statement_result = case_study_cacs_db.execute_select_statement(
    select_statement_text="""
        SELECT
            COUNT(DISTINCT wrpc.workbench_compound_id) AS number_of_workbench_reactions
        FROM workbench_reaction_product_compound AS wrpc
        WHERE wrpc.workbench_compound_id IN (
            SELECT
                DISTINCT workbench_compound_id
            FROM workbench_reaction_reactant_compound
        );
    """
).fetchone()

print((
    "Number of workbench chemical reactions with the product appearing as a reactant compound in other reactions: {number_of_workbench_reactions:,d}"
).format(
    number_of_workbench_reactions=select_statement_result[0]
))
```

# Chemical Reaction Transformation Patterns

```python
select_statement_result = case_study_cacs_db.execute_select_statement(
    select_statement_text="""
        SELECT
            COUNT(DISTINCT wrtp.workbench_reaction_pattern_id) AS number_of_workbench_reaction_transformation_patterns
        FROM workbench_reaction_transformation_pattern AS wrtp;
    """
).fetchone()

print(
    "Number of workbench chemical reaction transformation patterns: {number_of_workbench_reaction_transformation_patterns:,d}".format(
        number_of_workbench_reaction_transformation_patterns=select_statement_result[0]
    )
)
```

```python
select_statement_results = case_study_cacs_db.execute_select_statement(
    select_statement_text="""
        SELECT
            wrp.smarts AS workbench_reaction_transformation_pattern_smarts,
            COUNT(DISTINCT wrtp.workbench_reaction_id) AS number_of_workbench_reactions
        FROM workbench_reaction_transformation_pattern as wrtp
            LEFT JOIN workbench_reaction_pattern AS wrp ON wrtp.workbench_reaction_pattern_id = wrp.id
        GROUP BY wrp.smarts
        ORDER BY number_of_workbench_reactions DESC;
    """
).fetchall()

select_statement_data = dict()

print("The top ten most common chemical reaction transformation patterns:")

for select_statement_result_index, select_statement_result in enumerate(select_statement_results):
    if select_statement_result_index < 10:
        print(
            "{rank}. '{workbench_reaction_transformation_pattern_smarts:s}':{number_of_workbench_reactions:,d}".format(
                rank=select_statement_result_index+1,
                workbench_reaction_transformation_pattern_smarts=select_statement_result[0],
                number_of_workbench_reactions=select_statement_result[1]
            )
        )

    select_statement_data[select_statement_result[0]] = select_statement_result[1]
```

```python
visualization_dataframe = DataFrame(
    data=select_statement_data.values(),
    columns=[
        "number_of_workbench_reactions",
    ]
)

visualization_dataframe["number_of_workbench_reactions_per_transformation_pattern"] = range(1, len(visualization_dataframe) + 1)

visualization_dataframe["cumulative_percentage_of_workbench_reactions"] = (
    100 * visualization_dataframe["number_of_workbench_reactions"].cumsum() / sum(select_statement_data.values())
)

figure, axes_1 = pyplot.subplots(
    figsize=(14, 8),
    dpi=300
)

line_plot_1, = axes_1.plot(
    visualization_dataframe["number_of_workbench_reactions_per_transformation_pattern"],
    visualization_dataframe["number_of_workbench_reactions"],
    color="cornflowerblue",
    linewidth=2,
    label="Number of Workbench Chemical Reactions per Transformation Pattern"
)

axes_1.set_xlabel(
    xlabel="Number of Workbench Chemical Reaction Transformation Patterns (Linear Scale)",
    fontsize=12
)

axes_1.set_ylabel(
    ylabel="Number of Workbench Chemical Reactions (Log Scale)",
    fontsize=12
)

axes_1.set_yscale(
    value="log"
)

axes_1.set_xticks(
    ticks=range(0, len(visualization_dataframe) + 100000, 100000)
)

axes_1.set_xticklabels([
    "{tick:,}".format(
        tick=tick
    ) for tick in range(0, len(visualization_dataframe) + 100000, 100000)
])

axes_1.set_yticks(
    ticks=[10**i for i in range(0, 5)]
)

axes_1.set_yticklabels([
    "{tick:,}".format(
        tick=tick
    ) for tick in [10**i for i in range(0, 5)]
])

axes_1.yaxis.set_minor_locator(
    locator=LogLocator(
        base=10.0,
        subs=list()
    )
)

axes_1.yaxis.set_minor_formatter(
    formatter=NullFormatter()
)

axes_1.tick_params(
    axis="both",
    labelsize=12
)

axes_1.grid(
    visible=False
)

axes_2 = axes_1.twinx()

line_plot_2, = axes_2.plot(
    visualization_dataframe["number_of_workbench_reactions_per_transformation_pattern"],
    visualization_dataframe["cumulative_percentage_of_workbench_reactions"],
    color="firebrick",
    linewidth=2,
    label="Cumulative Percentage of Workbench Chemical Reactions per Transformation Pattern"
)

axes_2.set_ylabel(
    ylabel="Cumulative Percentage of Workbench Chemical Reactions (%)",
    fontsize=12
)

axes_2.set_ylim(
    bottom=0,
    top=105
)

axes_2.set_yticks(
    ticks=range(0, 110, 10)
)

axes_2.yaxis.grid(
    visible=True,
    which="major",
    linestyle="--",
    linewidth=0.5,
    alpha=0.7
)

axes_2.tick_params(
    axis="both",
    labelsize=12
)

cut_off_line = axes_1.axvline(
    x=visualization_dataframe.loc[
        visualization_dataframe[visualization_dataframe["cumulative_percentage_of_workbench_reactions"] >= 80].index[0],
        "number_of_workbench_reactions_per_transformation_pattern"
    ],
    color="black",
    linestyle="--",
    linewidth=1.5,
    label="80% of Workbench Chemical Reactions are Covered by {cut_off_value:,d} Transformation Patterns".format(
        cut_off_value=visualization_dataframe.loc[
            visualization_dataframe[visualization_dataframe["cumulative_percentage_of_workbench_reactions"] >= 80].index[0],
            "number_of_workbench_reactions_per_transformation_pattern"
        ]
    )
)

axes_1.legend(
    [line_plot_1, line_plot_2, cut_off_line],
    [line.get_label() for line in [line_plot_1, line_plot_2, cut_off_line]],
    loc="center right",
    fontsize=12
)

pyplot.tight_layout()

pyplot.show()
```
